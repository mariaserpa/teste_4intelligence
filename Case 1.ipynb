## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
from numpy import log
import matplotlib.pyplot as plt
import seaborn as sns;
sns.set()
from math import sqrt
import chardet
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_error
```

## Leitura da base e análise

```python
tfp = pd.read_csv("TFP.csv")
```

```python
tfp.head(10)
```

```python
tfp.isocode.unique()
```

## Criando novas bases

```python
usa = tfp[tfp.isocode == 'USA'].drop(columns= ['isocode']).set_index('year')
can = tfp[tfp.isocode == 'CAN'].drop(columns= ['isocode']).set_index('year')
mex = tfp[tfp.isocode == 'MEX'].drop(columns= ['isocode']).set_index('year')
```

```python
mediausa = []
stdusa = []
for item in range(len(usa)):
    media = usa['rtfpna'][:item].mean()
    std = usa['rtfpna'][:item].std()
    mediausa.append(media)
    stdusa.append(std)
    
usa['media'] = mediausa
usa['std'] = stdusa

mediacan = []
stdcan = []
for item in range(len(can)):
    media = can['rtfpna'][:item].mean()
    std = can['rtfpna'][:item].std()
    mediacan.append(media)
    stdcan.append(std)
    
can['media'] = mediacan
can['std'] = stdcan

mediamex = []
stdmex = []
for item in range(len(mex)):
    media = mex['rtfpna'][:item].mean()
    std = mex['rtfpna'][:item].std()
    mediamex.append(media)
    stdmex.append(std)
    
mex['media'] = mediamex
mex['std'] = stdmex
```

## Avaliação da estacionariedade

Nos dados estacionários, a média, variância e estrutura de autocorrelação não mudam no decorrer do tempo. É importante considerar a estacionariedade pois a maioria dos modelos leva em consideração a estacionariedade dos dados.

### Avaliando a distribuição dos dados (histograma)

```python
fig = plt.figure(figsize=(10.0, 10.0))

axes1 = fig.add_subplot(3, 1, 1)
axes2 = fig.add_subplot(3, 1, 2)
axes3 = fig.add_subplot(3, 1, 3)

axes1.hist(usa['rtfpna'])
axes1.title.set_text('USA')

axes2.hist(can['rtfpna'], label= 'TFP')
axes2.title.set_text('CAN')

axes3.hist(mex['rtfpna'], label= 'TFP')
axes3.title.set_text('MEX')

fig.tight_layout()
plt.show()
```

Obs.: Os dados não estão distribuidos uniformemente.

### Visualizando a serie temporal

```python
fig = plt.figure(figsize=(15.0, 10.0))

axes1 = fig.add_subplot(3, 1, 1)
axes2 = fig.add_subplot(3, 1, 2)
axes3 = fig.add_subplot(3, 1, 3)

axes1.plot(usa['rtfpna'], label= 'TFP')
axes1.plot(usa['media'], label='Média')
axes1.plot(usa['std'], label='Std')
axes1.title.set_text('USA')
axes1.legend()

axes2.plot(can['rtfpna'], label= 'TFP')
axes2.plot(can['media'], label='Média')
axes2.plot(can['std'], label='Std')
axes2.title.set_text('CAN')
axes2.legend()

axes3.plot(mex['rtfpna'], label= 'TFP')
axes3.plot(mex['media'], label='Média')
axes3.plot(mex['std'], label='Std')
axes3.title.set_text('MEX')
axes3.legend()

fig.tight_layout()
plt.show()
```

### Dividindo a base de treino e de teste

```python
anos = 15
treinousa = usa.iloc[:-anos,0:1].copy()
testeusa = usa.iloc[-anos:,0:1].copy()

treinocan = can.iloc[:-anos,0:1].copy()
testecan = can.iloc[-anos:,0:1].copy()

treinomex = mex.iloc[:-anos,0:1].copy()
testemex = mex.iloc[-anos:,0:1].copy()
```

```python
fig = plt.figure(figsize=(15.0, 10.0))

axes1 = fig.add_subplot(3, 1, 1)
axes2 = fig.add_subplot(3, 1, 2)
axes3 = fig.add_subplot(3, 1, 3)

axes1.plot(treinousa['rtfpna'], color='red')
axes1.plot(testeusa['rtfpna'], color='grey')
axes1.title.set_text('USA')

axes2.plot(treinocan['rtfpna'], color='red')
axes2.plot(testecan['rtfpna'], color='grey')
axes2.title.set_text('CAN')

axes3.plot(treinomex['rtfpna'], color='red')
axes3.plot(testemex['rtfpna'], color='grey')
axes3.title.set_text('MEX')

fig.tight_layout()
plt.show()
```

### Realizando o teste Dickey-Fuller (teste de hipótese sobre estacionariedade)

Considerando 99% de confiança, ou seja, p-value = 0.01. As séries (USA, CAN E MEX) não são estacionárias.

```python
usatest = treinousa['rtfpna'].values
result = adfuller(usatest)
print('Teste estatístico: %f' % result[0])
print('p-valaue: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
cantest = treinocan['rtfpna'].values
result = adfuller(cantest)
print('Teste estatístico: %f' % result[0])
print('p-valaue: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

### Transformando em estacionária (Método de diferenciação)

```python
diferusa = treinousa[['rtfpna']].diff().dropna()
fig = plt.figure(figsize=(15.0, 10.0))

plt.subplot(2,1,1)
plt.plot(diferusa)
plt.title('Primeira diferenciação - USA')

plt.subplot(2,1,2)
plt.hist(diferusa['rtfpna'])
plt.title('Nova distribuição - USA')

plt.show()

difusatest = diferusa['rtfpna'].values
result = adfuller(difusatest)
print('Teste estatístico: %f' % result[0])
print('p-valaue: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
